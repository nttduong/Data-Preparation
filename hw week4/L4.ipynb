**Exercise 1:** We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks. Combine them
into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises:

```python
#a. Read in the aapl.csv, amzn.csv, fb.csv, goog.csv, and nflx.csv files.
amzn = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/amzn.csv')
fb = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/fb.csv')
goog = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/goog.csv')
nflx = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/nflx.csv')
aapl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/aapl.csv')
```

```python
aapl.head()
```

```python
#b. Add a column to each dataframe, called ticker,indicating the ticker symbol it is for (Apple's is AAPL, forexample); 
#this is how you look up a stock. In this case, the filenames happen to be the ticker symbols.
amz = amzn.assign(ticker = 'AMZ')
fb = fb.assign(ticker = 'FB')
goog = goog.assign(ticker = 'GOOG')
nflx = nflx.assign(ticker = 'NFLX')
aapl = aapl.assign(ticker = 'AAPL')
```

```python
#c. Append them together into a single dataframe. (concat)
df = pd.concat([amz,fb,goog,nflx,aapl])
df.head()
```

```python
df.shape
```

```python
#d. Save the result in a CSV file called faang.csv. (to_csv)
df.to_csv('faang.csv')
```

**Exercise 2:** The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:


```python
#a. Read in the covid19_cases.csv file.
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/covid19_cases.csv')
data.head()
```

```python
data.shape
```

```python
#b. Create a date column using the data in the dateRep column and the pd.to_datetime() function.
data = data.assign(date= pd.to_datetime(data.dateRep))
data.head()
```

```python
#c. Set the date column as the index and sort the index.
data.set_index('date', inplace=True)
data.head()
```

```python
data.sort_index().head()
```

```python
#d. Replace all occurrences of United_States_of_America and United_ Kingdom with USA and UK, respectively. 
#Hint: the replace() method can be run on the dataframe as a whole.
data = data.replace('United_States_of_America', 'USA').replace('United_Kingdom', 'UK')
data.head()
```

```python
#e.Using the countriesAndTerritories column, filter the cleaned COVID-19 cases data down to
# Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA.
cleaned_data = data[data.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
cleaned_data.head()
```

```python
#f.Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). 
#Be sure to fill in NaN values with 0.

cleaned_data.reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
```
