```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/fb_stock_prices_2018.csv')
data.head()
```

```python
new = data.assign(week= pd.DatetimeIndex(data['date']).week)
new.head()
```

```python
data_dif = new.groupby(['week']).agg({"high":max,"low":min})
data_dif
```

```python
dif = data_dif['high'] - data_dif['low']
```

```python
plt.plot(dif)
plt.xlabel('week')
plt.ylabel('change')
plt.title('Facebook weekly high-low')
plt.show()
```

```python
fb = data.assign(change=data.high - data.low )
fb.head()
```

```python
plt.scatter(fb.volume,fb.change, c = 'b', s=30)
plt.xlabel('volume')
plt.ylabel('max_abs_change')
plt.title('Facebook Daily High - Low vs. Volume Traded')
plt.show()
```

```python
#Using the earthquake data, create box plots for the magnitudes of each magType used in Indonesia
eq = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/earthquakes.csv')
eq.head()
```

```python
mag_data= eq[['mag', 'magType']].groupby(['magType'])
mag_data
```

```python
mag_data.boxplot(figsize=(15, 8), subplots=False)
plt.title('Earthquake Magnitude Box Plots by magType')
plt.ylabel('magnitude')
```

```python
eq.boxplot(column='mag', by = 'magType', figsize=(15,8))
plt.title('Earthquake Magnitude Box Plots by magType')
```

```python
#Plot the bar chart displaying the number of earthquakes and tsunamis in Indonesia each day.
indo_eq = eq[eq.place.str.contains("Indonesia")]
indo_eq.head()
```

```python
indo_eq = indo_eq.assign(time=lambda x: pd.to_datetime(x.time, unit='ms'),earthquake=1)
indo_eq.head()
```

```python
new = indo_eq.set_index('time').resample('1D').sum()
new
```

```python
new.index = new.index.strftime('%b\n %d')
```

```python
new.plot(y=['earthquake', 'tsunami'], kind='bar',rot=0, figsize=(15,3), label=['earthquakes', 'tsunamis'],title='Earthquakes and Tsunamis in Indonesia ')
plt.xlabel('date')
plt.ylabel('count')
```
